```python
import requests
import defs
```

```python
session = requests.Session()
```

```python
instrument = "EUR_USD"
count = 10
granularity = "H1"
```

```python
url = f"{defs.OANDA_URL}/instruments/{instrument}/candles"
params  = dict(count=count, granularity=granularity, price = "MBA")
response = session.get(url, headers=defs.SECURE_HEADER, params=params)
```
