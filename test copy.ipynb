```python
import requests
import defs
import pandas as pd
```

```python
session = requests.Session()
```

```python
url = f"{defs.OANDA_URL}/accounts/{defs.ACCOUNT_ID}/instruments"
```

```python
url
```

```python
response = session.get(url, headers=defs.SECURE_HEADER, params=None)
```

```python
data = response.json()
```

```python
data.keys()
```

```python
instruments = data['instruments']
```

```python
len(instruments)
```

```python
instrument_data = []
for instrument in instruments:
    new_ob=dict(name= instrument['name'],
            type = instrument['type'],
            displayName = instrument['displayName'],
            pipLocation = instrument['pipLocation'],
            marginRate = instrument['marginRate']
            )
    instrument_data.append(new_ob)
```

```python
instrument_data
```

```python
instrument_df = pd.DataFrame.from_dict(instrument_data)
```

```python
instrument_df
```

```python
instrument_df.to_pickle("instrument.pkl")
```
